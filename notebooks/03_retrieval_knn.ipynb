# Dense KNN (TF-IDF + cosine) — Poda estrutural LLM

Usa `src/retrievers/dense.py`.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT))

from src.retrievers.dense import DenseRetriever
from src.preprocessing.config import PreprocessConfig
```

```python
pp_cfg = PreprocessConfig(use_stemming=False)
retriever = DenseRetriever(
    corpus_path=ROOT / "data/corpus.jsonl",
    preprocess_config=pp_cfg,
)
print(f"Indexed {len(retriever.doc_ids)} documents")
```

```python
queries = {
    "q1": "structural pruning of attention heads in transformer models",
    "q2": "LLM compression during fine-tuning",
    "q3": "MLP layer pruning for parameter efficiency",
}

for qid, qtext in queries.items():
    results = retriever.search(qtext, top_k=5)
    print(f"\n=== {qid}: {qtext} ===")
    for doc_id, score in results:
        print(f"  {doc_id}  {score:.4f}")
```

```python
run_path = ROOT / "eval/runs/knn.trec"
retriever.write_trec_run(queries, run_path, top_k=20, system_name="knn")
print(f"Run saved: {run_path}")
```
